# Friends Reference Lab — Visualization Client

This notebook only imports the installed package, loads CLI-generated artifacts, and displays summary tables and diagnostics. It does **not** implement tracking or reconciliation logic.

```python
import json
from pathlib import Path

from IPython.display import HTML, Image, display

repo_root = Path.cwd().parents[1]
run_dir = repo_root / "artifacts" / "video_reference"
# Resolve the newest run directory for the Friends video.
run_dirs = sorted(d for d in run_dir.rglob("raw/manifest.json") if d.exists())
manifest_path = run_dirs[-1].parent / "manifest.json" if run_dirs else None
manifest_path
```

```python
if manifest_path and manifest_path.exists():
    with open(manifest_path, "r", encoding="utf-8") as f:
        manifest = json.load(f)
    display(manifest)
else:
    print("No manifest found; run 'make video-reference-friends' first.")
```

```python
visual_dir = manifest_path.parent.parent / "visual"
for name in ["alignment_contact_sheet.jpg", "tracklet_contact_sheet.jpg", "canonical_contact_sheet.jpg", "quality_histograms.png", "timeline.png"]:
    path = visual_dir / name
    if path.exists():
        display(Image(filename=str(path)))
```

```python
report_path = manifest_path.parent.parent / "evaluation" / "final_report.json"
if report_path.exists():
    with open(report_path, "r", encoding="utf-8") as f:
        report = json.load(f)
    print("VERDICT:", report.get("verdict"))
    display(report.get("tracking", {}))
    display(report.get("reconciliation", {}))
```
